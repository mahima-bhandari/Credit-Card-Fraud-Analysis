```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['Class', 'datetime'], axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_train.value_counts()
```

## *Using SMOTE to handle class population imbalance*

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
x_train, y_train = smote.fit_resample(x_train, y_train)
x_test, y_test = smote.fit_resample(x_test, y_test)
```

```python
y_train.value_counts()
```

## *Logistic Regression for Classification*

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
```

```python
model.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
dff = pd.DataFrame(x)
feature_names = dff.columns
```

```python
coeff = model.coef_[0]
coeff
```

```python
sorted_indices = np.argsort(np.abs(coeff))[::-1]
sorted_coefficients = coeff[sorted_indices]
sorted_feature_names = feature_names[sorted_indices]
```

```python
# Print the sorted coefficients
print("Logistic Regression Coefficients:")
for feature, coef in zip(sorted_feature_names, sorted_coefficients):
    print(f"{feature}: {coef}")
```

```python
# Plot the sorted coefficients
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_coefficients)), sorted_coefficients)
plt.xticks(range(len(sorted_coefficients)), sorted_feature_names, rotation=90)
plt.xlabel('Features')
plt.ylabel('Coefficients')
plt.title('Logistic Regression Coefficients')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, pred)
confusion = confusion_matrix(y_test, pred)
precision = precision_score(y_test, pred)
recall = recall_score(y_test, pred)
f1 = f1_score(y_test, pred)
roc_auc = roc_auc_score(y_test, pred)
```

```python
tn, fp, fn, tp = confusion.ravel()
tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
ppv = precision
fpr = fp / (fp + tn)
fnr = fn / (fn + tp)
```

```python
# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(confusion)
print("Precision (PPV):", precision)
print("Recall (TPR):", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
print("True Positive Rate (TPR):", tpr)
print("True Negative Rate (TNR):", tnr)
print("False Positive Rate (FPR):", fpr)
print("False Negative Rate (FNR):", fnr)
```

## *False Negatve is 13.1% in Logistic Regression Model*

```python
# Plot the confusion matrix
plt.figure(figsize=(6, 6))
plt.imshow(confusion, cmap=plt.cm.Greens)
plt.title("Confusion Matrix for Logistic Regression Model")
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['CLASS 0', 'CLASS 1'])
plt.yticks(tick_marks, ['CLASS 0', 'CLASS 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
thresh = confusion.max() / 2.0
for i in range(confusion.shape[0]):
    for j in range(confusion.shape[1]):
        plt.text(j, i, format(confusion[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if confusion[i, j] > thresh else "black")
plt.tight_layout()
plt.show()
```

```python
# Plot the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(x_test)[:, 1])
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], 'w--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Logistic Regression Model')
plt.legend(loc="lower right")
plt.show()
```

```python
# Perform sensitivity analysis using permutation feature importance
from sklearn.inspection import permutation_importance
result = permutation_importance(model, x_test, y_test, n_repeats=10, random_state=42)
importances = result.importances_mean
std = result.importances_std
```

```python
# Sort the features by importance
sorted_idx = importances.argsort()[::-1]
sorted_importances = importances[sorted_idx]
sorted_std = std[sorted_idx]
sorted_feature_names = dff.columns[sorted_idx]
```

```python
# Print the sorted feature importances
print("Permutation Feature Importance:")
for feature, importance, std in zip(sorted_feature_names, sorted_importances, sorted_std):
    print(f"{feature}: {importance:.3f} (+/- {std:.3f})")
```

```python
# Plot the feature importances with error bars
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_importances)), sorted_importances, yerr=sorted_std, align='center', alpha=0.8, ecolor='black', capsize=10)
plt.xticks(range(len(sorted_importances)), sorted_feature_names, rotation=90)
plt.xlabel('Features')
plt.ylabel('Permutation Importance')
plt.title('Permutation Feature Importance for Logistic Regression Model')
plt.tight_layout()
plt.show()
```

## *KNN Algorithm* 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_values = list(range(1,11))
k_values
```

```python
accuracies = []
f1_scores = []
precisions = []
recalls = []
roc_auc_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    accuracies.append(accuracy)
    f1_scores.append(f1)
    precisions.append(precision)
    recalls.append(recall)
    roc_auc_scores.append(roc_auc)

best_k = k_values[np.argmax(roc_auc_scores)]
best_model = KNeighborsClassifier(n_neighbors=best_k)
best_model.fit(x_train, y_train)

y_pred_best = best_model.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred_best)
tn, fp, fn, tp = cm.ravel()
```

```python
accuracy_best = accuracy_score(y_test, y_pred_best)
precision_best = precision_score(y_test, y_pred_best)
recall_best = recall_score(y_test, y_pred_best)
f1_best = f1_score(y_test, y_pred_best)
roc_auc_best = roc_auc_score(y_test, y_pred_best)

tpr = tp / (tp + fn)
fpr = fp / (fp + tn)
tnr = tn / (tn + fp)
fnr = fn / (fn + tp)

print("Best K value:", best_k)
print("Accuracy:", accuracy_best)
print("Precision (PPV):", precision_best)
print("Recall (TPR):", recall_best)
print("F1 Score:", f1_best)
print("ROC AUC Score:", roc_auc_best)
print("True Positive Rate (TPR):", tpr)
print("True Negative Rate (TNR):", tnr)
print("False Positive Rate (FPR):", fpr)
print("False Negative Rate (FNR):", fnr)
```

## *False Negative is 17.84% in K Nearest Neighbor Algorithm*

```python
fpr_best, tpr_best, _ = roc_curve(y_test, best_model.predict_proba(x_test)[:,1])
plt.figure(figsize=(6,6))
plt.plot(fpr_best, tpr_best, label=f'ROC Curve (AUC = {roc_auc_best:.3f})')
plt.plot([0,1], [0,1], 'w--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for KNN model')
plt.legend(loc='lower right')
plt.show()
```

```python
# Plot the confusion matrix for the best model
plt.figure(figsize=(6, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix for KNN Model")
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['Class 0', 'Class 1'])
plt.yticks(tick_marks, ['Class 0', 'Class 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Add value labels inside each cell of the confusion matrix
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```

```python
# Plot ROC AUC vs K value
plt.figure(figsize=(8, 6))
plt.plot(k_values, roc_auc_scores, marker='o')
plt.xlabel('K Value')
plt.ylabel('ROC AUC')
plt.title('ROC AUC vs K Value')
plt.xticks(k_values)
plt.grid(True)

optimal_k_index = np.argmax(roc_auc_scores)
optimal_k = k_values[optimal_k_index]
optimal_roc_auc = roc_auc_scores[optimal_k_index]
plt.annotate(f'optimal k : {optimal_k}', xy = (optimal_k, optimal_roc_auc), xytext=(optimal_k + 0.6, optimal_roc_auc - 0.01),
             arrowprops=dict(facecolor='pink', shrink=0.6))
plt.show()
```
